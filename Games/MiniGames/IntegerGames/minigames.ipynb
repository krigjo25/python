```python
#   Importing Responsories
import sys
import random as r

from os import getenv
from dotenv import load_dotenv

#   Importing local libraries
from dictionary.gamedicitonary import GameOver, MathGame
from pylib.databasePython import MariaDB

class IntegerGames():

    '''
        #   Author : krigjo25
        #   Date   :  12.01-23

        #   Collection of Classic WordGames
    '''

    def GameLevel(self):

        '''
            #   Prompting the user for a level
            #   The level has to be greater than 0
        '''

        while True:

            try :

                #   Prompt a user for an integer
                lvl = int(input('level : '))

                #   Checking the condition of prompted message
                if lvl == None: raise TypeError('Input required')
                elif lvl < 0 : raise ValueError('Choose an integer grater than 0')

            except (ValueError, TypeError) as e:
                print(e)
                continue

            else: return lvl

    def GenerateIntegers(self, lvl):

        #   Matching the argument

        if lvl <= 10: return r.randrange(0, lvl * 10)
        elif lvl > 10 and lvl < 20:

            #   Match the level
            match lvl :
                case 11: return r.randrange(0, 200)
                case 12: return r.randrange(0, 300)
                case 13: return r.randrange(0, 400)
                case 14: return r.randrange(0, 500)
                case 15: return r.randrange(0, 600)
                case 16: return r.randrange(0, 700)
                case 17: return r.randrange(0, 800)
                case 18: return r.randrange(0, 900)
                case 19: return r.randrange(0, 1000)

        elif lvl >=20 and lvl < 30: return r.randrange(0, lvl * 10)


        return

    def Operators(self, lvl):

        #   Declare a list
        arg = []

        #   Initialize a dictionary
        dictionary = {  1:'+',
                        #2:'-',
                        #3:'/',
                        #4:'*',
                        }

        # Randomize the dictionary selection
        if "r" in lvl: r.shuffle(dictionary)
        n = r.randint(1, len(dictionary))

        #   Generate integers
        x = self.GenerateIntegers(lvl)
        y = self.GenerateIntegers(lvl)

        #   Match the dictionary value
        match dictionary[n]:

            case "+":
                x += y
                mf = f"{x} + {y} = "
            
            case "-":
                x -= y
                mf = f"{x} - {y} = "
        
            case '*':
                x *= y
                mf = f"{x} - {y} = "
        
            case'/':
                x /= y
                mf = f"{x} / {y} = "
        
            case '//':
                x //= y
                mf = f"{x} // {y} = "

        #   Append to list
        arg.append(x)
        arg.append(mf)

        #   Clear some memories
        del x, y
        del dictionary

        #   Returning the list
        return arg

    #   Games
    def GuessTheNumber(self):

        #   Initializing variables
        tempt = 5
        score = 0

        #   Game Configurations
        lvl = self.GameLevel()
        n = self.GenerateIntegers(lvl)


        while True:

            try :
                #   Prompting the user for an integer
                x = input('Guess: ').lower()

                #   Quit or raise value error
                if "q" in x : sys.exit(f"Game Summary\n Score : {score}\n {tempt} attempts left")
                elif x.isalpha(): raise ValueError("\"Q\" to quit, otherwise only digits allowed")

            except (ValueError, TypeError) as e: 

                print(f"{e}")
                continue

            else:

                x = int(x)
                print(n)
                #   Checking the conditions of x
                if x == n:

                    #   Increase score
                    score +=1

                    #   Print output
                    print(f"Score : {score} of 10 \n{MathGame().GuessTheNumber(n, x)}")
                    n = self.GenerateIntegers(lvl)

                else: 
                    print(MathGame().GuessTheNumber(n,x))

                #   End game
                if tempt == 0: return print(GameOver().IncorrectAnswer())
                elif score == 10: return print(GameOver().CorrectAnswer())

                #   Decrease the score by one
                tempt = -1

    def LittleProffessor(self):

        #   Game Configurations
        #   Prompting a level input
        lvl = self.GameLevel()

        #   Calculating the answer
        arg = self.Operators(lvl)

        #   Initializing variables
        score = 0
        tempt = 3

        while True:

            try :

                #   Prompting the user for the output
                prompt = input(f"{arg[1]}").lower()

                if "q" in prompt: sys.exit("Leaving the game. Thank you for playing")
                elif prompt.isalpha(): raise ValueError("\"Q\" to quit, otherwise digits is allowed")

            except ValueError as e:

                print('EEE')

            else:

                if prompt == arg[0]:

                    #   Adding one point to score
                    score += 1

                    #   Creating a new math problem to be solved
                    arg = self.Operators(lvl)

                    print(f"Just right!\nGame summuary\n Score: {score}\n Attempts left: {tempt}")


                else:

                    print('EEE')

                #   Breaking out of the loop
                if tempt <= 0:

                    print(f'Game summuary\n Score: {score}\n Attempts left: {tempt}\nCorrect number : {arg[0]} \n ')
                    break

                elif score == 9: 
                    print(f'Game summuary\n Score: {score}\n Attempts left: {tempt}')
                    
                    #   Clear some memory
                    del score, tempt
                    del arg

                    #   Break out of the loop
                    
            #   Decrease the variable
            tempt -= 1

            #   Clear some more memory
            del prompt, x, y

        return

if __name__ == '__main__':
    i = IntegerGames()
    i.GuessTheNumber()
```
